## Glosario

* artist: Artista de la canción (si la tupla representa un evento que implique la reproducción de una canción).

* auth: Autenticación ("Logged In" o "Logged Out").

* firstName: (Primer) Nombre del usuario.

* gender: Género del usuario.

* itemInSession: Número del item (registro) en la sesión del usuario.

* lastName: Apellido del usuario.

* length: "Longitud" de la canción.

* level: Tipo de suscripción ("free" o "paid").

* location: Ubicación del usuario.

* method: Método HTTP ligado al evento de usuario.

* page: Página que se muestra al usuario (UI).

* registration: Identificador único del registro (inicial) del usuario.

* sessionId: Identificador único de la sesión.

* song: Título de la canción (si el evento contiene).

* status: Código de estado HTTP (e.g. 200 "OK").

* ts: Timestamp Unix.

* userAgent: Agente del usuario (desde qué plataforma se ejecutó el evento).

* userId: Identificador único del usuario.

### 1. ¿Cuál es el top 5 de canciones?

```
top5Canciones = medium.groupBy("song").count().orderBy("count", ascending=False).filter(medium.song != "null")
display(top5Canciones)
```

### 2. ¿Cuál es el top 5 de artistas?

```
topArtist= medium.groupBy("artist").count().orderBy("count",ascending=False).filter(medium.artist!="null").take(5)
display(topArtist)
```

### 3. ¿Cuáles son las 5 canciones más largas?

```
top5CancionesLargas = medium.groupBy("song").avg("length").orderBy("avg(length)",ascending=False).take(5)
display(top5CancionesLargas)
```

### 4. ¿Cuáles son los 5 usuarios que escuchan más música?

```
top5Usuarios = medium.filter(medium.song != "null").groupBy("userId", "firstName", "lastName").count().orderBy("count", ascending=False).take(5)
display(top5Usuarios)
```

### 5. ¿Cuáles son los 5 usuarios que escuchan menos música?

```
bottom5Usuarios = medium.filter(medium.song != "null").groupBy("userId", "firstName", "lastName").count().orderBy("count", ascending=True).take(5)
display(bottom5Usuarios)
```

### 6. ¿Cuáles son las 5 ubicaciones en que más música se escucha?

```
top5Ubicaciones = medium.filter(medium.song != "null").groupBy("location").count().orderBy("count", ascending=False).take(5)
display(top5Ubicaciones)
```

### 7. ¿Cuáles son las 5 ubicaciones en que menos música se escucha?

```
bottom5Ubicaciones = medium.filter(medium.song != "null").groupBy("location").count().orderBy("count", ascending=True).take(5)
display(bottom5Ubicaciones)
```

### 8. ¿Quiénes escuchan más música: hombres o mujeres?

```
sexos = medium.filter(medium.song != "null").groupBy("gender").count().orderBy("count", ascending=True)
display(sexos)
```

### 9. ¿Cuál es el artista preferido por los hombres?

```
preferidoH =  medium.select(medium.gender, medium.artist).filter((medium.gender =="M")&(medium.artist  !="null")).groupBy("artist").count().orderBy("count",ascending = False).take(1)
display(preferidoH)
```

### 10. ¿Cuál es el artista preferido por las mujeres?

```
preferidoM =  medium.select(medium.gender, medium.artist).filter((medium.gender =="F")&(medium.artist  !="null")).groupBy("artist").count().orderBy("count",ascending = False).take(1)
display(preferidoM)
```

### 11. ¿Quién escucha más música: los usuarios pagos o los de versión gratuita?

```
usuarios = medium.filter(medium.song != "null").groupBy("level").count().orderBy("count", ascending=False)
display(usuarios)
```

### 12. De los usuarios que menos escuchan música, ¿cuáles de ellos usan la versión paga?

```
menos = medium.select(medium.firstName,medium.lastName,medium.level,medium.artist).filter((medium.level == "paid")&(medium.artist !="null")&(medium.firstName !="null")&(medium.lastName !="null")).groupBy("firstName","lastName","level").count().filter("count < 100").orderBy("count", ascending=True)
display(menos)
```

### 13. ¿Cuál es el promedio de tiempo (longitud) que escucha un usuario premium en una sesión?

```
promedio = medium.select(medium.firstName,medium.lastName,medium.level,medium.song,medium.length,medium.sessionId).filter((medium.level == "paid")&(medium.song !="null")&(medium.firstName !="null")&(medium.lastName !="null")).groupBy("firstName","lastName","level","sessionId").avg("length").orderBy("avg(length)")
display(promedio)
```

### 14. ¿Cuáles son los artistas más escuchados por usuarios premium que menos canciones escuchan?

```
top5Artistas_MenosEscuchas = medium.select(medium.firstName,medium.lastName,medium.level,medium.artist).filter((medium.level == "paid")&(medium.artist !="null")&(medium.firstName !="null")&(medium.lastName !="null")).groupBy("artist").count().filter("count < 100").orderBy("count", ascending=False)
display(top5Artistas_MenosEscuchas)
```

### 15. ¿Cuántos usuarios premium son hombres y cuántos son mujeres?

```
sexosPremium = medium.filter((medium.song != "null")&(medium.level == "paid")).groupBy("gender").count().orderBy("count", ascending=True)
display(sexosPremium)
```
